### **Hello, World! in Python**

If you're new to coding, you've probably heard of the "Hello, World!" program. It's the traditional first step when learning any programming language. Think of it like saying "hi" to the world of coding!

#### **What is it?**
A *"Hello, World!"* program simply prints the words Hello, World! to the screen. That's it! But behind those two little words is a huge milestone – we've just written and run our first piece of code.


#### **The Python Code**

````python
print("Hello, World!")
````

That's it! Just one line.

#### **How It Works**
`print()` is a built‑in Python function that outputs text (or other data) to the console.

The text we want to print goes inside the parentheses, surrounded by quotation marks `(" ")`. Quotation marks tell Python: *"This is a string of text, not a command."*

#### **How to Run It**
1. Save the code in a file called hello.py (the .py extension tells your computer it's a Python file).

2. Open a terminal (command prompt) in the folder where you saved hello.py.

3. Run the command:
```bash
python hello.py
```

4. You should see:
```txt
Hello, World!
```

#### **Try Experimenting**
Change the text inside the quotes:

````python
print("Good morning, Sir/Madam!")
````

Or print multiple things:

````python
print("Hello", "World", "from Python!")
````
